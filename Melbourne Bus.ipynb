```python
def fix_geometries(council_graph):
    """Adds a direct line geometry for a street wherever it is absent from the data"""
    x_values = nx.get_node_attributes(council_graph, 'x')
    y_values = nx.get_node_attributes(council_graph, 'y')
    # We will work with the edges and add the missing geometries (lines denoting the roads between points)
    graph_with_geometries = list(council_graph.edges(data=True))
    # Iterate through the edges and, where missing, add a geometry attribute with the line between start and end nodes
    for e in graph_with_geometries:
        if not 'geometry' in e[2]:
            #print('n', end='')
            e[2]['geometry'] = LineString([Point(x_values[e[0]], y_values[e[0]]),
                                           Point(x_values[e[1]], y_values[e[1]])])
    #graph_with_geometries[0:4]
    # We will now drop the start and end nodes, as we will construct a new Spaghetti network based on the geometries of the roads
    road_lines = [x[2] for x in graph_with_geometries]
    # From this, we can construct a GeoDataFrame
    return gpd.GeoDataFrame(pd.DataFrame(road_lines))

def ratio_from_intersection(point):
    dists = list(distance_dict[point.stop_id].values())
    length = sum(dists)
    return min(dists[0]/length, dists[1]/length)

def dist_from_intersection(point):
    dists = list(distance_dict[point.stop_id].values())
    return min(dists[0], dists[1]) * 1000

def stop_colour(stop):
    if stop.inter < 0.1:
        return 'r'
    elif stop.inter < 0.3:
        return 'b'
    else:
        return 'g'
```
