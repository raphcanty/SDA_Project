```python
import osmnx
import geopandas as gpd
from shapely.geometry import Point, LineString, Polygon
import pandas as pd
```

```python
london_stops = gpd.read_file('first_outputs/london_processed.shp')
london_stops
```

```python
melb_stops = gpd.read_file('first_outputs/melbourne_processed.shp')
melb_stops
```

```python
melbourne_polygon = \
    osmnx.geocode_to_gdf('Melbourne, Victoria, Australia').geometry[0]
```

```python
london_polygon = \
    osmnx.geocode_to_gdf('London, England, United Kingdom').geometry[0]
london_polygon = Polygon(london_polygon.exterior)
```

```python
glen_iris = osmnx.geocode_to_gdf('Glen Iris, Melbourne, Victoria, Australia').geometry[0]
```

## Melb Diversity Process

```python
melb_pois = osmnx.geometries_from_polygon(melbourne_polygon, tags={'amenity': True, 'shop': True,
                                                                   'leisure': True, 'healthcare': True})
melb_pois['geometry'] = melb_pois.geometry.apply(lambda x: x.centroid)
melb_pois = melb_pois[['unique_id', 'amenity', 'shop', 'healthcare', 'leisure', 'geometry', 'name']]
melb_pois
```

```python
print('+' * (len(melb_stops)//100))
```

```python
melb_data = []
count = 0
for stop in melb_stops.iloc:
    count += 1
    if count == 100:
        print('+', end='')
        count = 0
    zone = stop.geometry.buffer(0.0005)
    inside = melb_pois[melb_pois.within(zone)]
    melb_data.append([stop.stop_id, inside.amenity.count(), inside.shop.count(),
                      inside.healthcare.count(), inside.leisure.count()])
print('done')
```

```python
melb_processed = pd.DataFrame(melb_data, columns=['stop_id', 'amenity', 'shops', 'healthcare', 'leisure'])
melb_processed.to_csv('melb_diversity.csv')
```
